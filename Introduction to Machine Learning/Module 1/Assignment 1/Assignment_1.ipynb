### Days

### Pop

In this dataframe, I wanted to keep it simple by using just pop and the city. I used a for loop to iterate through the cities,
as well as change the Municipality to CITY and 'Population as of 2019' to pop for the purpose of preparing for the join early
on. 


```python
df_pop = df.loc[df['Municipality'] == 'Rio Branco']
df_pop = df_pop.loc[df_pop['State'] == 'AC']
cities = ['Rio Branco', 'Maceió', 'Macapá', 'Manaus', 'Salvador', 'Fortaleza', 'Brasília', 'Vitória', 'Goiânia', 'São Luís', 'Cuiabá', 'Campo Grande', 'Belo Horizonte', 'Belém', 'João Pessoa', 'Curitiba', 'Recife', 'Teresina', 'Rio de Janeiro', 'Natal', 'Porto Alegre', 'Porto Velho', 'Boa Vista', 'Florianópolis', 'São Paulo', 'Aracaju', 'Palmas']

for i in range(1,27):
    df_pop = df_pop.append(df.loc[df['Municipality'] == cities[i]])

df_pop = df_pop[['Municipality', 'Population as of 2019']]
df_pop.columns = [i.replace('Municipality','CITY') for i in df_pop.columns]
df_pop.columns = [i.replace('Population as of 2019','pop') for i in df_pop.columns]
df_pop = df_pop.drop_duplicates()
df_pop = df_pop.drop([6512,6516, 6901, 6331, 6909, 9400]).sort_values(by=['CITY'])
df_pop
```

# First merging

Next, I used the IS_CAPITAL list to formulate my list of which were actually the capital cities in quesiton. Then after
replacing the ANNUAL with temp, I merged the prior two dataframes df_pop and result_df with the titles <b> Pop </b> and <b> Days </b>. 

```python
df2["IS_CAPITOL"]=df2["IS_CAPITOL"].replace('y','Y')
df2 = df2.loc[df2['IS_CAPITOL'] == 'Y']
result_df2 = df2[['CITY', 'ANNUAL']]
result_df2.columns = [i.replace('ANNUAL','temp') for i in result_df2.columns]
result_df2.sort_values(by=['CITY'])
joined_df_pop_temp = df_pop.merge(result_df2, how='inner', left_on='CITY', right_on = 'CITY')
joined_df_pop_temp.columns = [i.replace('Population as of 2019','pop') for i in joined_df_pop_temp.columns]
joined_df_pop_temp_days = result_df.merge(joined_df_pop_temp, how='inner', left_on='Municipality', right_on = 'CITY')
joined_df_pop_temp_days
```

```python
# Here I imported the third dataframe from the excel sheet with the city area. 

df3 = pd.read_excel(xls, 'City area')
df3.columns = [i.replace('City','CITY') for i in df3.columns]
```

### Rearranging the order and dropping columns

Here I am rearranging the order as well as dropping the columns to prepare for features: days_cube, days_sq, days, temp, pop_dense_sq, pop_dense, pop.

```python
column_names = ["days_cube", "days_sq", "days", "temp", "pop_dense_sq", "pop_dense", "pop"]
#df = df.reindex(columns=column_names)

joined_df2 = df3.merge(joined_df_pop_temp_days, how='inner', left_on='CITY', right_on = 'CITY')
joined_df2['pop'] = joined_df2['pop'].astype('float')
joined_df2['SQ_KM'] = joined_df2['SQ_KM'].astype('float')
joined_df2['pop_dense'] = joined_df2['pop']/joined_df2['SQ_KM']
joined_df2['pop_dense_sq'] = joined_df2['pop_dense']**2
joined_df2['days_sq'] = joined_df2['days']**2
joined_df2['days_cubed'] = joined_df2['days']**3
```

### Accumulated Cases
